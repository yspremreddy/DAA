```python
import time
def iterative_sum(n):
    result = 0
    for i in range(1, n + 1):
        result += i
    return result

def recursive_sum(n):
    if n == 1:
        return 1
    else:
        return n + recursive_sum(n - 1)

```

```python
import matplotlib.pyplot as plt
def measure_time(func, n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time
N_values = range(1, 1001)

iterative_times = [measure_time(iterative_sum, n) for n in N_values]
recursive_times = [measure_time(recursive_sum, n) for n in N_values]

plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (s)')
plt.title('Time Taken to Compute Sum of First N Natural Numbers')
plt.legend()
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

def generate_random_array(size):
    return [random.randint(1, 1000) for _ in range(size)]
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

def measure_time(func, arr, key):
    start_time = time.time()
    func(arr, key)
    end_time = time.time()
    return end_time - start_time

array_size = 10000
random_array = generate_random_array(array_size)
search_keys = [random.randint(1, 1000) for _ in range(5)]
repetitions = 100
linear_times = []
binary_times = []

for key in search_keys:
    linear_time_total = 0
    binary_time_total = 0
    for _ in range(repetitions):
        linear_time_total += measure_time(linear_search, random_array, key)
        binary_time_total += measure_time(binary_search, sorted(random_array), key)
    linear_times.append(linear_time_total / repetitions)
    binary_times.append(binary_time_total / repetitions)
plt.plot(range(1, 6), linear_times, label='Linear Search')
plt.plot(range(1, 6), binary_times, label='Binary Search')
plt.xlabel('Search')
plt.ylabel('Time (s)')
plt.title('Average Time Taken by Linear and Binary Search Algorithms (100 repetitions)')
plt.legend()
plt.show()
```

```python
def string_to_integer(s):
    if not s:
        return 0
    
    digit = int(s[0])

    rest_value = string_to_integer(s[1:])
    
    return digit * (10 ** len(s[1:])) + rest_value

input_string = "13531"
result = string_to_integer(input_string)
print("String:", input_string)
print("Integer:", result)
```

```python
def reverse_string(s):
    
    if len(s) <= 1:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

input_string = "pots&pans"
reversed_string = reverse_string(input_string)
print("Original string:", input_string)
print("Reversed string:", reversed_string)
```

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    else:
        return s[0] == s[-1] and is_palindrome(s[1:-1])

# Example usage:
input_string1 = "racecar"
input_string2 = "gohangasalamiimalasagnahog"
print("Is", input_string1, "a palindrome?", is_palindrome(input_string1))
print("Is", input_string2, "a palindrome?", is_palindrome(input_string2))
```
